```python
import requests,json,numpy as np, datetime,matplotlib.pyplot as plt 
```

```python
alamat = "https://api.covid19api.com/total/country/Indonesia/status/confirmed"
#?from=2020-03-01T00:00:00Z&to=2020-04-01T00:00:00Z
parameter = {
    'from':"2020-03-01T00:00:00Z",
    'to' : "2020-04-01T00:00:00Z"
}
```

```python
response = requests.get(alamat, params = parameter)
```

```python
text = response.json()
```

```python
text
```

```python
result = [data['Cases'] for data in text]
date = [datetime.datetime.strptime(data['Date'],'%Y-%m-%dT%H:%M:%SZ').strftime('%Y-%m-%d') for data in text]
```

```python
print(result)
print(date)
```

```python
plt.figure(figsize=(30,10),dpi=120)
plt.style.use('seaborn-whitegrid')
plt.plot(date,result)
plt.xticks(rotation=90)
plt.show()
```
